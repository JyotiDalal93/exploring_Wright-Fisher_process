# A population evolving as per the "Wright-Fisher process" (https://en.wikipedia.org/wiki/Genetic_drift and http://www.math.ubc.ca/~db5d/SummerSchool09/lectures-dd/lectures6-7.pdf) with two types/alleles (A and a) present at t = 0 (or initially) #

## Cases: ##
    
### 1) Both the types, A and a have same fitness, i.e., genetic drift is the only evolutionary force acting on the population. Initial frequency of allele A = 0.5 for all cases. ###

### 1.1.) Population size = 10 ###

```python
import matplotlib.pyplot as plt
import numpy
```

```python

# Initial frequency of allele A = 0.5, in all the populations 

data = list(numpy.loadtxt("WF_pop size = 10, no selection.txt", comments = '#', usecols = (2)))

data1 = data[0:31]       # allele frequency in 1st population 
data2 = data[31:62]      # allele frequency in 2nd population 

data3 = data[62:93]     # allele frequency in 3rd population 
data4 = data[93:124]    # allele frequency in 4th population 

data5 = data[124:155]   # allele frequency in 5th population 
data6 = data[155:186]   # allele frequency in 6th population 

data7 = data[186:217]   # allele frequency in 7th population 
data8 = data[217:248]   # allele frequency in 8th population 

data9 = data[248:279]    # allele frequency in 9th population 
data10 = data[279:310]   # allele frequency in 10th population 

x_p = [i for i in range(31)] # time (in generations) for each population to evolve

plt.ylim([0, 1.1])

plt.xlabel('time (generations)')
plt.ylabel('Frequency of allele A')
plt.title('Population size = 10, ONLY genetic-drift is present')

plt.plot(x_p, data1)
plt.plot(x_p, data2)
plt.plot(x_p, data3)
plt.plot(x_p, data4)
plt.plot(x_p, data5)
plt.plot(x_p, data6)
plt.plot(x_p, data7)
plt.plot(x_p, data8)
plt.plot(x_p, data9)
plt.plot(x_p, data10)

plt.hlines(0.5, 0, 30, linestyles = 'dotted' )

data = list(numpy.loadtxt("WF_pop size = 1000, no selection.txt", comments = '#', usecols = (2)))

data1 = data[0:31]       # allele frequency in 1st population 
data2 = data[31:62]      # allele frequency in 2nd population 

data3 = data[62:93]     # allele frequency in 3rd population 
data4 = data[93:124]    # allele frequency in 4th population 

data5 = data[124:155]   # allele frequency in 5th population 
data6 = data[155:186]   # allele frequency in 6th population 

data7 = data[186:217]   # allele frequency in 7th population 
data8 = data[217:248]   # allele frequency in 8th population 

data9 = data[248:279]    # allele frequency in 9th population 
data10 = data[279:310]   # allele frequency in 10th population 

x_p = [i for i in range(31)] # time (in generations) for each population to evolve

plt.ylim([0, 1.1])

plt.xlabel('time (generations)')
plt.ylabel('Frequency of allele A')
plt.title('Population size = 1000, ONLY genetic-drift is present')

plt.plot(x_p, data1)
plt.plot(x_p, data2)
plt.plot(x_p, data3)
plt.plot(x_p, data4)
plt.plot(x_p, data5)
plt.plot(x_p, data6)
plt.plot(x_p, data7)
plt.plot(x_p, data8)
plt.plot(x_p, data9)
plt.plot(x_p, data10)

plt.hlines(0.5, 0, 30, linestyles = 'dotted' )
```

In populations of size 10, either allele A is fixed (frequency = 1) or lost (frequency = 0).

### 1.2.) Population size = 1000 ###

```python
# Initial frequency of allele A = 0.5, in all the populations 

data = list(numpy.loadtxt("WF_pop size = 1000, no selection.txt", comments = '#', usecols = (2)))

data1 = data[0:31]       # allele frequency in 1st population 
data2 = data[31:62]      # allele frequency in 2nd population 

data3 = data[62:93]     # allele frequency in 3rd population 
data4 = data[93:124]    # allele frequency in 4th population 

data5 = data[124:155]   # allele frequency in 5th population 
data6 = data[155:186]   # allele frequency in 6th population 

data7 = data[186:217]   # allele frequency in 7th population 
data8 = data[217:248]   # allele frequency in 8th population 

data9 = data[248:279]    # allele frequency in 9th population 
data10 = data[279:310]   # allele frequency in 10th population 

x_p = [i for i in range(31)] # time (in generations) for each population to evolve

plt.ylim([0, 1.1])

plt.xlabel('time (generations)')
plt.ylabel('Frequency of allele A')
plt.title('Population size = 1000, ONLY genetic-drift is present')

plt.plot(x_p, data1)
plt.plot(x_p, data2)
plt.plot(x_p, data3)
plt.plot(x_p, data4)
plt.plot(x_p, data5)
plt.plot(x_p, data6)
plt.plot(x_p, data7)
plt.plot(x_p, data8)
plt.plot(x_p, data9)
plt.plot(x_p, data10)

plt.hlines(0.5, 0, 30, linestyles = 'dotted' )
```

In populations of size 1000, neither allele A is fixed (frequency = 1) nor lost (frequency = 0). After 30 generations, both the alleles are present in all 10 populations.

NOTE: **As population size increases, the variance in the allele frequencies/strength of genetic-drift decreases.**

While the *mean* remains same, i.e., 0.5.

# Exploring fixation probability of allele A in the presence of various evolutionary forces #

Fixation probability = probability with which an allele takes over the whole population.


## Some results: ##


### A) Selection is ABSENT ###

If s = 0.0, fixation probability of an allele is simply equal to its initial frequency. 

$$ \pi = \frac{\text{Initial number of allele A}}{\text{population size}} $$



Using,

$N$ for population size, $s$ for selection coefficient, $i$ for the initial number of allele A.


*When $N = 10,$ $i = 1$*

$\pi$ (expected) = 0.1 and *$\pi$ (simulations) = 0.1008* 


*When $N = 10,$ $i = 8$*

$\pi$ (expected) = 0.8 and *$\pi$ (simulations) = 0.801* 





### B)  Selection is also present ### 

Some special cases:

#### B.1) $2 \times s << 1$ and $2 \times N \times s << 1$####

If the conditions given above are satisfied, then 

$$ \pi = \frac{\text{Initial number of allele A}}{\text{population size}} $$



### Examples ###



B.1.1) When $N = 10,$ $i = 1$, s = 0.001

     2*N*s = 0.02 << 1 
     2*s = 0.002 << 1


$\pi$ (simulations) = 0.1003 $\approx$ $\frac{1}{10} = 0.1.$


#### B.2) $2 \times s << 1$ and $2 \times N \times s >> 1$####

If the conditions given above are satisfied, then 

$$ \pi = 2*s $$


### Examples ###

B.2.1) When $N = 1000,$ $i = 1$, s = 0.005

     2*N*s = 10 >> 1 
     2*s = 0.010 << 1


$\pi$ (simulations) = 0.0112 $\approx$ 2*s = 0.010.


For a better understanding of fixation probability of an allele under various evolutionary forces, refer to the classical paper by *Motoo Kimura* http://www.physics.rutgers.edu/~morozov/677_f2017/Physics_677_2017_files/Kimura_Genetics1962_classic.pdf
